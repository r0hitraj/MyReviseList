### Print the sum , diffrence, product and quotient of two number

```python
num1, num2 = map(int , input("Enter The Numbers").split()) # GIVE A IMPUT THEN GIVE A SPACE THEN AGAIN NEXT INPUT
# for quotient use // and for exact division use /
sum = num1+num2
diff1 = num1-num2
quotient1 = num1//num2
quotient2 = num2//num1
print(f"Sum {sum}")
print(f"Diffrence: {diff1}")
print(f"quotient1 {num1} / {num2} = {quotient1}")
print(f"quotient2 {num2} / {num1} = {quotient2}")
```

### Find the square and cube of a number

```python
num = int(input("Enter the Number"))
square = num**2
cube = num **3
print(f"Square of {num}= {square}")
print(f"Cube of {num}= {cube}")
```

### Swap two numbers using a temporary variable

### Find the remainder when one number is divided by another

### Calculate area and perimeter of a rectangle

```python
length,breadth = map (int , input("Enter Lenght, Breadth").split())
perimeter = (length + breadth) * 2
area = length * breadth
print(f"Perimeter : {perimeter}")
print(f"Area: {area}")
```

### check if a year is a leap year.

```python
#what is leap year if a year is divisible by 4 and not divisible by 100 or it sould be divisible by 400
year = int(input("Enter the year"))
print("Leap year") if (year % 4 ==0 or year %100 !=0) or (year %400 == 0) else print("Normal Year")
```

### check whether a character is a vowel or consonant

```python
vowel = "aeiouAEIOU"
ch = str(input("Enter a letter"))
print("Vowel") if ch in vowel else print("Consonant")
#=====================================================
#haveing saftey , another version of same script
#=====================================================

if len(ch) == 1 and ch.isalpha():
    print("Vowel" if ch in vowel else "Consonant")
else:
    print("Invalid input")
```
